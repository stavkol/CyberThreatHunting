```python
# φόρτωση βιβλιοθηκών
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
%matplotlib inline

# έλεγχος κατανομής κλάσεων
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 14, 8
RANDOM_SEED = 42
LABELS = ["Normal", "Fraud"]

# φόρτωση συνόλου δεδομένων
url = "https://raw.githubusercontent.com/nsethi31/Kaggle-Data-Credit-Card-Fraud-Detection/master/creditcard.csv"
data = pd.read_csv(url)

# διαγραφή κενών ή μη αριθμητικών τιμών
data = data.select_dtypes(include=['float64', 'int64']).dropna()

# διαχωρισμός ανεξάρτητων και εξαρτημένων μεταβλητών
# The target variable in this dataset is 'Class' not 'Frequency'
y=data['Class']
X=data.drop(['Class'], axis=1)


# διαχωρισμός συνόλου δεδομένων σε σύνολο εκπαίδευσης και δοκιμών
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# κατανομή της εξαρτημένης μεταβλητής (class)
count_classes = pd.value_counts(data['Class'], sort = True)
count_classes
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
#εκπαίδευση αλγορίθμου
rf_classifier = RandomForestClassifier(n_estimators=100)
rf_classifier.fit(X_train, y_train)
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)
# υπολογισμός μέτρων αξιολόγησης
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
accurracy_score = accuracy_score(y_test, y_test_pred)
print("Accuracy Score : ", accurracy_score)
precision = precision_score(y_test, y_test_pred)
print("Precision Score : ", precision)
recall = recall_score(y_test, y_test_pred)
print("Recall Score : ", recall)
f1 = f1_score(y_test, y_test_pred)
print("F1 Score : ", f1)
confusion_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix : ", confusion_matrix)
```
